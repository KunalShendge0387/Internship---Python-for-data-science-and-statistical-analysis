```python
# Hypothesis test

'''
Research Question - Is the number of males and females equal in organisation?

Sample - 46
Male - 34
Female - 12

Null Hypothesis (H0) - There is no difference in the number of male and female in the organisation
Alternative Hypothesis (Ha) - There is a difference in the number of male and female in the organisation

Two-sided test - non directional (equal or unequal)
One-sided test - greater or lesser (directional)
'''
```

```python
from scipy.stats import binom_test

binom_test(x = 34, n =  46, p = 0.5)
#x - count of a category
#n - sample size
#p - probability to belong to any category
```

```python
p_value = 0.0016414913408482334

if p_value < 0.05:
    print('Reject the null hypothesis')
else:
    print('Fail to reject the null hypothesis')
```

```python
# Null Hypothesis is rejected

# There is a difference in the number of male and female in the organisation.
```
