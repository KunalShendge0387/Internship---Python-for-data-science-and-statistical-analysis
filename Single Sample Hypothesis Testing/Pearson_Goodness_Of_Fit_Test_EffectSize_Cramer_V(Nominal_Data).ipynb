```python
import pandas
data = pandas.read_csv("GSS2012a.csv")
data
```

```python
data['mar1'].value_counts()
```

```python
'''
Research Question - Are people more likely to get married?

H0 - There is no preference and the distribution of each category is same
Ha - There is a preference and the distribution of each category is not same
'''
```

```python
from scipy.stats import chisquare

chisquare(data['mar1'].value_counts())
```

```python
statistic=1249.1262235960844
pvalue=3.564167265021484e-269
```

```python
# Cramer V effect size
# n - samples
# df - number of categories - 1

n = data.shape[0]
df = 4
cv = (statistic / (n * df)) ** 0.5
cv
```

```python
# if df is k

'''
0 to 0.1/sqrt(k-1) - negligible
0.1/sqrt(k-1) to 0.3/sqrt(k-1) - small
0.3/sqrt(k-1) to 0.5/sqrt(k-1) - medium
0.5/sqrt(k-1) and above - large
'''
```

In the population the different categories are not evenly distributed as per the pearson chisquare goodness of fit test (p<0.05) and the difference in the categories is ..... (cramer v = 0.397)
