```python
import pandas
import numpy

num_patients = 20

before = numpy.random.randint(1, 5, size = num_patients)
after = numpy.random.randint(1, 5, size = num_patients)

print(before)
print(after)
```

```python
data = pandas.DataFrame({'before': before, 'after': after})
data
```

```python
crosstab = pandas.crosstab(data['before'], data['after'])
crosstab
```

```python
from scipy.stats import wilcoxon

wilcoxon(data['before'], data['after'], correction=False)
```

```python
statistic=47.5
pvalue=0.7503158528871918
```

```python
from scipy.stats import norm

z_value = norm.ppf(1 - (pvalue/2))

import math

r_ros = abs(z_value / math.sqrt(data.shape[0]))

r_ros
```

```python
if r_ros < 0.2:
  print('Very low')
elif r_ros < 0.4:
  print('Low')
elif r_ros < 0.6:
  print('Moderate')
elif r_ros < 0.8:
  print('Strong')
else:
  print('Very Strong')
```
